```python
import matplotlib.pyplot as plt
```

```python
x =[10,50,20,80]
y=[10,20,30,40]
plt.plot(x,y)
plt.xlabel('EJE X')
plt.ylabel('EJE Y')
plt.title('EJEPLO DE LINEAS')
plt.show()
```

```python
x1 = ['A','B','C','D']
y1 = [100,50,80,40]
plt.xlabel('LETRAS')
plt.ylabel('NUMEROS')
plt.title('EJEPLO DE B')
plt.bar(x1,y1)
plt.show()
```

```python
x2= ['A','B','C','D']
y2 = [100,50,80,40]
plt.pie(y2, labels=x2, autopct = '%1.1f%%')
plt.title('EJEMPLO DE PIE O PASTEL')
plt.axis('equal') # SIEMPRE SEA CIRCULAR
plt.show()
```

```python
x3 = [100,50,80,40]
y3= [10,50,20,40]
z3 = [18,50,20,100]
fig = plt.figure()
graf = fig.add_subplot(111, projection='3d')
graf.plot(x3,y3,z3 ,  color='green')
plt.title('EJEMPLO DE 3D')
graf.set_xlabel('EJE X')
graf.set_ylabel('EJE Y')
graf.set_zlabel('EJE Z')
plt.show()
```

```python
from matplotlib import cm
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
x4 = [100,50,80,40]
y4= [10,50,20,40]
x4,y4 = np.meshgrid(x4,y4)
z4 = np.array([
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
    [18,50,20,100],
])
fig = plt.figure()
graf = fig.add_subplot(111, projection='3d')
aux = graf.plot_surface(x4,y4,z4 ,cmap=cm.viridis)
fig.colorbar(aux)
plt.show()
```
